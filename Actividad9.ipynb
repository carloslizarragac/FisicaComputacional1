### Teoría de Estabilidad de Sistemas de Ecuaciones Diferenciales Ordinarias Autónomas.

Estudiaremos los problemas de valor inicial 

\begin{equation*}
\frac{d \vec{x}}{d t} = A \vec{x}, \ \mathrm{con} \ \vec{x}(0)=\vec{x_0}
\end{equation*}

donde $A$ es una matriz cuadrada constante de dimensión $n \times n$ 

Si $n=1$, la solución de la ecuación diferencial es 

\begin{equation*}
\vec{x}(t) = x_0 \exp{(At)}
\end{equation*}

Si $n > 1$, la solución también tendrá la misma forma, pero con $A$ siendo una matriz. 

Podemos definir la exponencial de una matriz cuadrada constante A de dimensiones $n \times n$

\begin{equation*}
\exp(A) = \sum_{k=0}^{\infty} \frac{A^k}{k!}
\end{equation*}

Esto es similar al desarrollo en Serie de Taylor de $\exp(x)$, alrededor de $x_0=0$.

Vimos en el tema de Álgebra Lineal, que si la matriz es diagonalizable (todos sus eigenvalores son distintos), entonces existe una matriz $P$ formada por los eigenvectores de $A$, tal que $A=P^{-1} \Lambda P$, donde $\Lambda$ es la matriz diagonal formada por los $n$ eigenvalores $(\lambda_1, \lambda_2, \ldots, \lambda_n)$. 

También vimos que el cálculo de potencias de la matriz $A$ se simplificaban computacionalmente

\begin{equation*}
A^k = P^{-1} \Lambda^k P
\end{equation*}

Si podemos diagonalizar la matriz $A$ ($n$ eigenvalores distintos), y $P = [ x_1 x_2  \ldots  x_n ]$ es un eigenvector de $A$, tendremos

\begin{equation*}
A = [x_1 x_2 \ldots x_n]^{-1} \ \begin{bmatrix}
\lambda_1 & 0 & \cdots & 0 \\
0 & \lambda_2 & \cdots & 0 \\
\vdots & \vdots &      & \vdots \\
0 & 0 & \cdots & \lambda_n \\
\end{bmatrix} \ [x_1 x_2 \ldots x_n]   
\end{equation*}

por lo que entonces se tiene que 

\begin{equation*}
At = [x_1 x_2 \ldots x_n]^{-1} \ \begin{bmatrix}
\lambda_1 t & 0 & \cdots & 0 \\
0 & \lambda_2 t & \cdots & 0 \\
\vdots & \vdots &      & \vdots \\
0 & 0 & \cdots & \lambda_n t \\
\end{bmatrix} \ [x_1 x_2 \ldots x_n]   
\end{equation*}

y calculando la exponencial de $\exp(At)$ sería equivalente a calcular

\begin{equation*}
\exp(At) = [x_1 x_2 \ldots x_n]^{-1} \ \begin{bmatrix}
\exp(\lambda_1 t) & 0 & \cdots & 0 \\
0 & \exp(\lambda_2 t) & \cdots & 0 \\
\vdots & \vdots &      & \vdots \\
0 & 0 & \cdots & \exp(\lambda_n t) \\
\end{bmatrix} \ [x_1 x_2 \ldots x_n]  
\end{equation*}

Es fácil ver el comportamento de la solución cuando  $\lim t \rightarrow  \infty$, lo determinan las exponenciales de los eigenvalores. 


**Sistema de ecuaciones lineales homogéneas.**

Supongamos que tenemos el siguiente sistema de ecuaciones diferenciales

\begin{eqnarray*}
\frac{d x_1}{dt} & = & a x_1 + b x_2 \\
\frac{d x_2}{dt} & = & c x_1 + d x_2 \\
\end{eqnarray*}

Si $\vec{x}^T = (x_1, x_2)$, entonces la ecuación anterior se puede escribir como 

\begin{equation*}
\frac{d \vec{x}}{dt} = \frac{d}{dt} \begin{bmatrix}
x_1 \\ x_2 \end{bmatrix} = A \begin{bmatrix}
x_1 \\ x_2 \end{bmatrix}, \mathrm{donde} \ A = \begin{bmatrix}
a & b \\ c & d \end{bmatrix}
\end{equation*}

Si definimos como $p = \mathrm{Tr}\ (A) = (a+d)$ y $q=\det (A) = (ad-bc)$

El polinomio característico de $\det(A - \lambda I)$ satisface

\begin{equation*}
p(\lambda) = \lambda^2 - p \lambda + q  
\end{equation*}

cuyas raíces son

\begin{equation*}
\lambda_{1,2} = \frac{p \pm \sqrt{p^2 - 4 q}}{2}
\end{equation*}

donde el discriminante $\Delta = p^2 - 4q$ puede ser mayor o menor que cero (es decir, $\lambda$  Real o Compleja).

Se puede ver que $p$ y $q$ satisfacen:

\begin{eqnarray*}
p & = & \lambda_1 + \lambda_2 \\
q & = & \lambda_1 \cdot \lambda_2 \\
\end{eqnarray*}





**Clasificación de las soluciones**

Para sistema de matrices 2x2, podemos tener 3 posibles categorías:

\begin{eqnarray*}
A) \begin{bmatrix}
\lambda & 0 \\
0 & \mu
\end{bmatrix}; B) \begin{bmatrix}
\lambda & 0 \\
1 & \lambda
\end{bmatrix}; C) \begin{bmatrix}
a & -b \\
b & a
\end{bmatrix}, b\neq0
\end{eqnarray*}

Estas 3 categorías pueden dividirse en 14 casos de puntos críticos:

**Caso A)**

\begin{equation*}
 \begin{bmatrix}
\lambda & 0 \\
0 & \mu
\end{bmatrix}
\end{equation*}

La solución se puede escribir como 

\begin{equation*}
x(t) = c_1 \exp(\lambda t) \ {v_1} + c_2 \exp(\mu t) \ {v_2}
\end{equation*}

donde ${v_1}$ y ${v_2}$ son los eigenvectores correspondentes a los eigenvalores $\lambda$ y $\mu$ respectivamente, $c_1$ y $c_2$ constantes.

*Puntos críticos:* 

1.   Punto silla: $\lambda < 0 < \mu$
2.   Nodo estable: $\lambda < \mu < 0$
3.   Nodo estable: $\lambda=\mu < 0$
4.   Nodo inestable: $0 < \lambda < \mu$
5.   Nodo inestable: $0 < \lambda = \mu$
6.   Nodo degenerado: $\lambda < \mu =0$
7.   Nodo degenerado: $0 = \mu < \lambda$
8.   Nodo degenerado: $0 = \mu = \lambda$

**Caso B)**

\begin{equation*}
 \begin{bmatrix}
\lambda & 0 \\
1 & \lambda
\end{bmatrix}
\end{equation*}


La solución se puede escribir como 

\begin{equation*}
x(t) = [ (c_1 + c_2 t) \ v_1 + c_2 v_2 ] \ \exp(\lambda t)
\end{equation*}

donde $c_1$ y $c_2$ son constantes y los vectores $v_1$ y $v_2$ satisfacen las ecuaciones

\begin{eqnarray*}
A v_1 & = & \lambda v_1 \\
A v_2 & = & \lambda v_1 + v_2
\end{eqnarray*}

*Puntos críticos:* 

9.  Nodo estable: $\lambda < 0$
10. Nodo inestable: $0 < \lambda$
11. Nodo degenerado: $\lambda = 0$

**Caso C)**

\begin{equation*}
 \begin{bmatrix}
a & -b \\
b & a
\end{bmatrix}, b \neq 0
\end{equation*}

El polinomio característico es 

\begin{eqnarray*}
(a-\lambda)^2 + b^2 & = & 0 \\
\lambda^2 + a^2 - 2 a \lambda + b^2 & = & 0 \\
\lambda^2 - 2 a \lambda + (a^2+b^2) & = & 0 \\  
\end{eqnarray*}

cuyas raíces son complejas:

\begin{equation*}
\lambda_{1,2} = a \pm i b
\end{equation*}

La solución se puede escribir como 

\begin{eqnarray*}
x(t) & = & c_1 \exp((a+ib)t) \ \ v_1 + c_2 \exp((a-ib)t) \ v_2 \\
& = & c_1 \exp(at) \exp(ibt) \ v_1 +
c_2 \exp(at) \exp(-ibt) \ v_2 \\
& = & [c_1 \exp(ibt) \ v_1 +
c_2 \exp(-ibt) \ v_2] \ \exp(at) \\
& = & [c_1 (\cos(bt) + i \sin(bt)) \ v_1 +
c_2 (\cos(bt) - i \sin(bt)) \ v_2 ] \exp(at) \\
\end{eqnarray*}

donde $c_1$ y $c_2$ son constantes complejas y los vectores $v_1$ y $v_2$ son vectores complejos en $C^2$.

*Puntos críticos:*

12. Foco o Espiral estable: $a < 0 < b$
13. Foco o Espiral inestable: $b < 0 < a$
14. Centro: $a=0, b>0$
